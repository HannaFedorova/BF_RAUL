Добавим столбец с длительностью трудоустройства для каждого кандидата:

```python
raul_df['длительность_трудоустройства'] = raul_df['завершение_трудоустройства'] - raul_df['начало_трудоустройства']
```

Получили столбец в формате данных Timedelta. Приведем формат к integer:

```python
raul_df['длительность_трудоустройства'] = pd.to_numeric(raul_df['длительность_трудоустройства'].dt.days, downcast='integer')
```

Напишем функцию для распределения кандидатов по группам длительности трудоустройства:

```python
def длительность (row):
        if row['длительность_трудоустройства'] < 1:
            return 'меньше одного дня'
        if row['длительность_трудоустройства'] >= 1 and row['длительность_трудоустройства']<=30:
            return 'от одного дня до месяца'
        if row['длительность_трудоустройства'] > 30 and row['длительность_трудоустройства']<=182:
            return 'от одного месяца до полугода'
        return 'больше полугода'
               
raul_df['длительность'] = raul_df.apply(длительность, axis = 1)
```

```python
def от_одного_дня (row):
        if row['длительность_трудоустройства'] > 1:
            return 'истина'
        return 'ложь'
               
raul_df['от_одного_дня'] = raul_df.apply(от_одного_дня, axis = 1)
```

```python
def от_одного_месяца (row):
        if row['длительность_трудоустройства'] > 30:
            return 'истина'
        return 'ложь'
               
raul_df['от_одного_месяца'] = raul_df.apply(от_одного_месяца, axis = 1)
```

```python
def от_трех_месяцев (row):
        if row['длительность_трудоустройства'] > 91:
            return 'истина'
        return 'ложь'
               
raul_df['от_трех_месяцев'] = raul_df.apply(от_трех_месяцев, axis = 1)
```

```python
def от_полугода (row):
        if row['длительность_трудоустройства'] > 182:
            return 'истина'
        return 'ложь'
               
raul_df['от_полугода'] = raul_df.apply(от_полугода, axis = 1)
```
